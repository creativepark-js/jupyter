```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
    cat = ''
    if age <= 5 :
        cat = 'Baby'
    elif age <= 12:
        cat = 'Child'
    elif age <= 18:
        cat = 'Teenager'
    elif age <= 25:
        cat = 'Student'
    elif age <= 35 :
        cat = 'Young Adult'
    elif age <= 60 :
        cat = 'Adult'
    else :
        cat = 'Elderly'
    
    return cat
```

```python
df['Age_cat'] = df['Age'].apply(lambda x : get_category(x))
df[['Age','Age_cat']].head(5)
```
